# 크롤링

```python
url = 'https://www.nsmall.com/'
#headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser', from_encoding='utf-8')
soup1 = list(soup.children)[3]
body1 = list(soup1.children)[3]
```

## 카테고리, 카테고리별 링크

```python
C1_len = len(list(body1.select('div.category_list_wrap > h3')))

result = []
for i in range(C1_len) :
    # 대분류 이름 저장
    name1 = list(body1.select('div.category_list_wrap > h3'))[i].text
    
    # 대분류 안의 카테고리로 들어가자
    tmp1 = list(body1.select('div.category_list_wrap > ul'))[i]
    # 대분류 안에 중분류 몇 개인지 count
    C2_len = len(list(tmp1.select('li.cate_item')))
    
    for j in range(C2_len) :
        # 중분류 이름 저장
        name2 = list(tmp1.select('li.cate_item > a > span'))[j].text
        
        # 중분류 안의 카테고리로 들어가자
        tmp2 = list(tmp1.select('li.cate_item > div.low_depth_wrap > ul'))[j]
        # 중분류 안에 소분류 몇 개인지 count
        C3_len = len(list(tmp2.select('li')))
        
        for k in range(C3_len) :
            # 소분류 저장
            name3 = list(tmp2.select('li > a > span'))[k].text

            # 링크 저장용
            tmp3 = list(tmp2.select('li'))[k]
            link = str(list(tmp3.select('a'))[0]).split("'")[1].replace('amp;','')
            
            # 결과 저장
            result.append([name1, name2, name3, link ])
            
df_category = pd.DataFrame(result, columns=['C1', 'C2', 'C3', 'link'])
df_category
```

## 카테고리별 브랜드

```python
brand_list = []
for link in df_category['link'] :
    new_url = url + link
    req = requests.get(new_url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser', from_encoding='utf-8')
    soup1 = list(soup.children)[3]
    body1 = list(soup1.children)[3]
    
    
    row_len = len(list(body1.select('#brandResultList > #result1 > p')))
    brand_name = []
    for row in range(row_len) : 
        x = list(body1.select('#brandResultList > #result1 > p'))[row]
        span_len = len(x.select('span.txt'))

        for span in range(span_len) :
            xx = x.select('span.txt')[span]
            brand_name.append(xx.text)
            
    brand_list.append(brand_name)
```

```python
df_category['brand'] = brand_list
df_category
```
