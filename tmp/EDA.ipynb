```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np
from datetime import datetime
import time

# 한글 폰트 설정

fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/NanumGothic.ttf' # For Windows
font_name = fm.FontProperties(fname=font_location).get_name()
plt.rc('font', family=font_name)
plt.rc('font', size=20)

# 지수 표현 안보기
```

```python
# 데이터 로드
data = pd.read_csv('data/train.csv', encoding='cp949', header=0, parse_dates=["방송일시"])

# 상품군 = 무형 제거 

data.drop(data[data['상품군']=='무형'].index, inplace=True)
data.reset_index(inplace=True)


# 판매단가, 취급액 쉼표 제거 , int 변환

data['판매단가'] = data['판매단가'].str.replace(',','').astype(int)
data[' 취급액 '] = data[' 취급액 '].str.replace(',','').astype(int)
data['노출(분)'] = data['노출(분)'].astype(int)


# 방송일시 column 분리 ( , 일, 시, 분, 요일 )

data['방송일시'] = pd.to_datetime(data['방송일시'], format='%Y-%m-%d %H:%M:%S', errors='raise')

data['month'] = data['방송일시'].dt.month
data['day'] = data['방송일시'].dt.day
data['hour'] = data['방송일시'].dt.hour
data['minute'] = data['방송일시'].dt.minute
data['요일'] = data['방송일시'].dt.weekday

# 06 ~ 익일 02 가 한 cycle임을 고려하여 0, 1, 2 에 +24를 해주자

# def plus_24(x) :
#     if x in [0, 1, 2] :
#         return x + 24
#     else :
#         return x
    
# data['hour'] = list(map(plus_24, data['hour']))


overnight_index_list = data.loc[(data["hour"] >=0) & (data["hour"] < 3)].index

overnight_index_list
data.loc[overnight_index_list, "hour"] += 24
data.loc[overnight_index_list, "day"] -= 1
data.loc[overnight_index_list, "요일"] -= 1
data.loc[data["요일"] == -1, "요일"] = 6

# 엑셀에서 수정하자
# month 홀수의 새벽은 day = 31로, 요일은 전날로
```

```python
# use data
df = data.loc[:,['month','day','hour','minute','노출(분)','상품코드','상품명','판매단가',' 취급액 ','요일','상품군','마더코드']]
df
```

```python
def replace_word(x) :
    if '무)'in x :
        return x.replace('무)', '무이자')
    elif '일)' in x :
        return x.replace('일)', '일시불')
    else :
        return x
    
df['상품명'] = list(map(replace_word, df['상품명'])) 
```

```python
new = []
for i in range(len(df['상품명'])) :
    if '일시불' in df['상품명'][i] :
        df['상품명'][i] = df['상품명'][i].replace('일시불','')
        new.append(1)
    elif '무이자' in df['상품명'][i] :
        df['상품명'][i] = df['상품명'][i].replace('무이자','')
        new.append(2)
    else :
        new.append(0)

df['pay_type'] = pd.DataFrame(new)
```

```python
df.to_csv('./analysis/분석용2.csv', index=False, encoding='cp949')
# 엑셀에서 월, 일 주기 맞춰주기 !
```

## 각 요일별 상품군의 취급액 평균(mean), 편성 개수 (count)

```python
clrs = ['c','m','y','b','lime','violet','g']

def cnt_product_day(day, color) :
    plt.figure(figsize=(20,5))
    plt.rc('font', size=20)
    df[df['요일']==day].groupby(df['상품군']).count()[' 취급액 '].plot(kind='bar', color=color)
    plt.ylim = [0,1250]
    plt.title(day)
    plt.show()
    
for day, color in zip(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], clrs) :
    cnt_product_day(day, color)
```

```python
def mean_product_day(day, color) :
    plt.figure(figsize=(20,5))
    plt.rc('font', size=20)
    df[df['요일']==day].groupby(df['상품군']).mean()[' 취급액 '].plot(kind='bar', color=color)
    plt.title(day)
    plt.ylim = [0,6]
    plt.show()
    
for day, color in zip(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], clrs) :
    mean_product_day(day, color)
```

```python
plt.figure(figsize=(20,5))
plt.rc('font', size=20)   
for day, color in zip(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], clrs) :
    df[df['요일']==day].groupby(df['상품군']).mean()[' 취급액 '].plot(color=color)
```

## 각 상품군별 요일의 취급액 평균(mean), 편성 개수 (count)


```python
def mean_day_category(category_name) :
    plt.figure(figsize=(20,5))
    plt.rc('font', size=20)
    df[df['상품군']==category_name].groupby(df['요일']).mean()[' 취급액 '].sort_values(ascending=False).plot(kind='bar')
    plt.title(category_name)

for category in df['상품군'].unique() :
    mean_day_category(category)
```

```python
def count_day_category(category_name) :
    plt.figure(figsize=(20,5))
    plt.rc('font', size=20)   
    df[df['상품군']==category_name].groupby(df['요일']).count()[' 취급액 '].sort_values(ascending=False).plot(kind='bar')
    plt.title(category_name)
    plt.ylim = [0,5]

for category in df['상품군'].unique() :
    count_day_category(category)
```

```python
def make_new_col(word) :
    new = []
    for i in range(len(df['상품명'])) :
        if word in df['상품명'][i] :
            df['상품명'][i] = df['상품명'][i].replace(word,'')
            new.append(1)
        else :
            df['상품명'][i] = df['상품명'][i].replace(word,'')
            new.append(0)
    df[word] = pd.DataFrame(new)
    return df

#make_new_col('무이자')
#make_new_col('일시불')
#make_new_col('햇')
#make_new_col('초특가')
#make_new_col('국내산')
#make_new_col('무료')
```
