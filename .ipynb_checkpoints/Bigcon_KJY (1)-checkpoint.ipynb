```python
dum1 = pd.get_dummies(df['상품군'],prefix = '상품군')
df2 = pd.concat([df, dum1], axis = 1)
```

```python
import datetime as dt
```

```python
df2['방송일시'] = pd.to_datetime(df2['방송일시'], format='%Y-%m-%d %H:%M:%S', errors='raise')
```

```python
df2['date']        = df['방송일시'].dt.date 
df2['월'] = df['방송일시'].dt.month_name()        # 월(문자)
df2['월_숫자'] = df['방송일시'].dt.month          # 월(숫자)

df2['일']        = df['방송일시'].dt.day          # 일(숫자)

df2['시']       = df['방송일시'].dt.hour          # 시(숫자)

df2['분']     = df['방송일시'].dt.minute          # 분(숫자)
```

```python

df2['요일']  = df2['방송일시'].dt.day_name()
df2['요일_숫자']  = df2['방송일시'].dt.weekday            # 요일이름(문자) (=day_name())

df2['weekofyear']    = df2['방송일시'].dt.weekofyear      # 연 기준 몇주째(숫자) (=week)
df2['dayofyear']     = df2['방송일시'].dt.dayofyear       # 연 기준 몇일째(숫자)

df2['month_start']    = df2['방송일시'].dt.is_month_start # 월 시작일 여부
df2['month_end']     = df['방송일시'].dt.is_month_end     #월 마지막일 여부
df2['year_start']    = df['방송일시'].dt.is_year_start    # 연 시작일 여부
df2['year_end']      = df['방송일시'].dt.is_year_end      # 연 마지막일 여부

```

#### onehot 인코딩

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()

enc.fit(np.array(df2['month_start']).reshape(-1,1))
df2['month_start'] = enc.transform(np.array(df2['month_start']).reshape(-1,1)).toarray()

enc.fit(np.array(df2['month_end']).reshape(-1,1))
df2['month_end'] = enc.transform(np.array(df2['month_end']).reshape(-1,1)).toarray()

enc.fit(np.array(df2['year_start']).reshape(-1,1))
df2['year_start'] = enc.transform(np.array(df2['year_start']).reshape(-1,1)).toarray()

enc.fit(np.array(df2['year_end']).reshape(-1,1))
df2['year_end'] = enc.transform(np.array(df2['year_end']).reshape(-1,1)).toarray()



```

# 모델

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
```

```python
X = data.drop(['취급액'],axis = 1)
y = data['취급액']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 777)


rfc = RandomForestRegressor(random_state = 777)


param ={
    'n_estimators'      : [20],
    'max_depth'         : [20,40],
    'min_samples_leaf'  : [4,10,15],
    'min_samples_split' : [10,100,1000],
    'bootstrap' : [True]
}

grid_rfc = GridSearchCV(rfc, param_grid = param)

grid_rfc.fit(X_train, y_train)

print('최적 파라미터      : ', grid_rfc.best_params_)
print('최고 예측 정확도   : {0:.4f}'.format(grid_rfc.best_score_))

estimator = grid_rfc.best_estimator_

pred = estimator.predict(X_test)
print('검증 데이터 정확도 : {0:.4f}'.format(r2_score(y_test, pred)))
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 

    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y_test, pred)
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

#피처 중요도 도표 그리기
featureImportance = estimator.feature_importances_
#가장 높은 중요도 기준으로 스케일링
featureImportance = featureImportance/featureImportance.max()
sorted_idx = np.argsort(featureImportance[:22])
barPos = np.arange(sorted_idx[:22].shape[0])+ .5
plot.barh(barPos, featureImportance[sorted_idx[:22]], align='center')
plot.yticks(barPos, X_train.keys()[sorted_idx][:22])
plot.xlabel('Variable Importance')
plot.show()
```
