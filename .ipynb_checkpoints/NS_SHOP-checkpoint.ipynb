# NS Shop+ 판매 실적 예측

-실적데이터 : 19년1월~19년12월 프로그램별 실적 데이터  
-외부데이터 : 요일별/시간대별 분단위 시청률 데이터 (단위 %)  
-예측시점 : 20년6월 프로그램별 분당실적 예측  
-평가방법 : MAPE(Mean Absolute Percentage Error)  

```python
import pandas as pd
import numpy as np
```

## Load dataset

```python
train = pd.read_csv("./data/train.csv", encoding='cp949', header=0, parse_dates=["방송일시"])
test = pd.read_csv("./data/test.csv", header=1, parse_dates=["방송일시"])
print("train shape:", train.shape)
print("test shape:", test.shape)
```

## Preprocessing

```python
train.drop(train.loc[train["상품군"] == "무형"].index, inplace=True)
test.drop(test.loc[test["상품군"] == "무형"].index, inplace=True)
print("train shape:", train.shape)
print("test shape:", test.shape)
```

```python
# 결측치 체크
train.isnull().sum()
```

```python
# column에 있는 공백 제거
train.columns = train.columns.str.strip()
train.columns
```

```python
# 판매단가와 취급액을 'int64' 타입으로 변경
train["판매단가"] = train["판매단가"].str.replace(",","").astype("int64")
train["취급액"] = train["취급액"].str.replace(",","").astype("int64")
```

```python
train.dtypes
```

```python
# 방송일시를 월,일,시간,요일로 변경하여 저장
train["방송일시"] = pd.to_datetime(train["방송일시"], format='%Y-%m-%d %h:%m', errors='raise')
train["방송일시_month"] = train["방송일시"].dt.month
train["방송일시_day"] = train["방송일시"].dt.day
train["방송일시_hour"] = train["방송일시"].dt.hour
train["방송일시_minute"] = train["방송일시"].dt.minute
train["day_of_the_week"] = train["방송일시"].dt.dayofweek
train.head()
```

```python
train.drop("방송일시", axis=1, inplace=True)
train
```

```python
# 홈쇼핑 하루 방송 사이클이 오전 06시(06시20분)~익일 02시(02시20분) 까지 이므로 날짜를 맞춰주는 작업 수행
# 0시~3시 사이의 방송시간을 24시~27시로 변경
# 방송 날짜와 요일을 하루전으로 변경
overnight_index_list = train.loc[(train["방송일시_hour"] >=0) & (train["방송일시_hour"] < 3)].index
overnight_index_list
train.loc[overnight_index_list, "방송일시_hour"] += 24
train.loc[overnight_index_list, "방송일시_day"] -= 1
train.loc[overnight_index_list, "day_of_the_week"] -= 1
train.loc[train["day_of_the_week"] == -1, "day_of_the_week"] = 6
```

```python
train.tail()
```

```python
# 다음연도로 넘어간 항목은 따로 날짜를 변경
new_year_january_list = [38300, 38301, 38302, 38303]
train.loc[new_year_january_list, "방송일시_month"] = 12
train.loc[new_year_january_list, "방송일시_day"] = 31
```

```python
train.tail()
```

```python
# 상품군 one-hot encoding
train_dummy = pd.get_dummies(train, columns=['상품군'], prefix='상품군', drop_first=True)
```

```python
# 상품코드를 기준으로 데이터를 묶고 '노출(분)', '판매단가', '취급액', '마더코드', 상품군(one-hot)을 평균값으로 계산
# '판매단가', '마더코드는', 상품군은 평균을 내도 기존값 그대로 나옴
# 취급액, 노출 sum으로 바꾸기
train_grouped_by_product_code = train_dummy.pivot_table(index='상품코드',
                                                  values=['노출(분)', '판매단가', '취급액', '마더코드',
                                                         '상품군_가전', '상품군_건강기능', '상품군_농수축',
                                                          '상품군_생활용품', '상품군_속옷', '상품군_의류',
                                                          '상품군_이미용', '상품군_잡화', '상품군_주방', '상품군_침구'], aggfunc=np.mean)
```

## MAPE

```python
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
rf_model.fit(X_train, y_train)
pred = rf_model.predict(X_test)
```

```python
pred = np.exp(pred)
y_test = np.exp(y_test)
```
